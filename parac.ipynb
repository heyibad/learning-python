```python
import pandas as pd
import numpy as np

# Load the dataset
data_path = "googleplaystore.csv"  # Replace with your actual file path
df = pd.read_csv(data_path)

# 1. App Category Popularity
category_counts = df['Category'].value_counts()

# Option 1: Print top 10 categories
print("Top 10 App Categories:")
print(category_counts.head(10))

# Option 2: Visualize with bar chart (using matplotlib or seaborn)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
category_counts.plot(kind="bar", color="skyblue")
plt.title("Distribution of Apps by Category")
plt.xlabel("App Category")
plt.ylabel("Number of Apps")
plt.xticks(rotation=45)
plt.tight_layout()
# plt.show()  # Uncomment to display the plot

# 2. The Rating Rollercoaster
ratings = df["Rating"]

# Option 1: Print descriptive statistics
print("\nDescriptive Statistics of Ratings:")
print(ratings.describe())

# Option 2: Visualize with histogram
plt.figure(figsize=(8, 5))
plt.hist(ratings, bins=20, edgecolor="black")
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Apps")
plt.grid(axis="y")
plt.tight_layout()
# plt.show()  # Uncomment to display the plot

# 3. Reviews and Ratings: Hand in Hand?
reviews = df["Reviews"]

# Calculate correlation coefficient (Pearson's r)
correlation = np.corrcoef(ratings, reviews)[0, 1]

print("\nCorrelation between Ratings and Reviews:", correlation)

# Optionally, visualize with scatter plot (using matplotlib or seaborn)
# ... (code similar to histogram plot can be used)

# Save plots (optional)
# plt.savefig("category_distribution.png")
# plt.savefig("rating_distribution.png")

# Further analysis (replace with your desired code)
# ...
```
